Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.


How many such routes are there through a 20×20 grid?

```python
def grid_pathfinder(x,y):
    if x == 0 and y == 0: return 1
    num_paths = 0
    
    if x > 0:
        num_paths += grid_pathfinder(x-1, y)
    if y > 0:
        num_paths += grid_pathfinder(x, y-1)
        
    return num_paths
```

```python
print grid_pathfinder(2,2)
```

```python
print grid_pathfinder(20,20)
```

The above method is too slow for larger grids
